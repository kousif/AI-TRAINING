#training

```python
# Create the data.yaml file
yaml_content = """
train: /content/data/train/images
val: /content/data/test/images # Corrected valid to test path
test: /content/data/test/images

nc: 1  # Number of classes (assuming only 'signboard')
names: ['signboard']  # Class names
"""

with open("data.yaml", "w") as f:
    f.write(yaml_content)

print("âœ“ data.yaml created successfully!")
```

```python
from ultralytics import YOLO
model = YOLO('yolov8s.pt')

'''Here, a YOLOv8 small model (pretrained on COCO dataset) is loaded.
This pretrained model acts as the base that will be fine-tuned using our custom signboard dataset.'''


model.train(
    data='data.yaml', # This line is already correct
    epochs=10,
    imgsz=50,
    batch=16,
    name='signboard-detector'
)

'''This command fine-tunes YOLOv8 on our dataset.

data='data.yaml' tells YOLO where the dataset is and how itâ€™s structured.

epochs=50 means the model will train for 50 full passes through the dataset.

imgsz=640 resizes all images to 640Ã—640 pixels.

batch=16 means 16 images will be processed at a time.

name='signboard-detector' gives a custom name to this training run, so all outputs are saved under that name.'''
```

# testing/prediction

```python
# List the contents of the runs directory to find the correct training folder name
!ls /content/runs/detect/
```

```python
import cv2
import os
from ultralytics import YOLO
import glob


# Update the model path to the correct training run directory
model = YOLO('/content/runs/detect/signboard-detector2/weights/best.pt')

'''This loads the best model obtained after training (based on validation performance).

We then create a folder to save cropped signboard images:'''

os.makedirs('crops', exist_ok=True)


img_paths = glob.glob('/content/data/test/images/1828.png')

'''This loads one test image (1828.png) from the test folder.

Now for each image, we make predictions and crop detected signboards:'''

for img_path in img_paths:
    results = model.predict(source=img_path, save=False)
    img = cv2.imread(img_path)
    base = os.path.basename(img_path).split('.')[0]

    for i, box in enumerate(results[0].boxes):
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        crop = img[y1:y2, x1:x2]
        cv2.imwrite(f'crops/{base}_crop_{i}.jpg', crop)


        '''Hereâ€™s what happens:

The model predicts objects (signboards) in the image.

results[0].boxes contains bounding boxes for each detection.

(x1, y1, x2, y2) are coordinates of the detected region.

The detected area is cropped from the original image using OpenCV.

Each cropped signboard is saved in the crops/ folder with a unique name.'''
```
